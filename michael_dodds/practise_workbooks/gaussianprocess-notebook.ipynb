We are going to work with GPs in exactly the same setting as in the case of linear regression. This means we are interested in the task of observing a data-set $\mathcal{D} = \{x_i, y_i \}_{i=1}^N$ where we assume the following relationship between the variates,  
$$
y_i = f(x_i).
$$
Our task is to infer the function $f(\cdot$) from $\mathcal{D}$. We will make the same assumptions of independence as in the linear regression case which leads us to the following likelihood function,
$$
p(\mathbf{y}\vert f(\mathbf{x})) = \prod_{i=1}^N p(y_i \vert f(x_i)),
$$
where $f(x_i)$ is the output of the function at location $x_i$.

**Question 1**
Explain what the independence assumption implies? How would the likelihood look if we could not make this assumption?/

**Answer**
The likelihood function gives us the probability of seeing the data given $f(x_i)$, implying that $y_i$ does not give any information about $y_i$, in other words the evaluation of the function at $x_i$ does not depend on the evaluation of the function at any other $x_i$. In the case that $y_i$ gives information about $y_i$ the joint likelihood would no longer factorize into the product of the individual probabilities $y_i$ given $f(x_i)$ . The new probability would be given by $p(y\vert f(x)) = p(y_1, y_2, ..., y_N \vert f(x))$. A covariance term would have to account for the relation of all $y_i$.

### Prior



We will now proceed to define the prior over the function values $p(\mathbf{f})$, we will do so using a Gaussian process. The prior being non-parametric needs in addition to a set of parameters $\boldsymbol{\theta}$ also access to the location where we want to evaluate it $\mathbf{x}$. This means the distribution that we are going to parametrise is,
$$
p(\mathbf{f}\mid \mathbf{x},\boldsymbol{\theta}) = \mathcal{N}\left(\mathbf{f}\mid \mu(\mathbf{x}, k(\mathbf{x},\mathbf{x}) \right).
$$
We will begin by implementing the co-variance function $k(x_i,x_j)$. We will play around with



```python
def rbf_kernel(x1, x2, varSigma, lengthscale):
    if x2 is None:
        d = cdist(x1, x1)
    else:
        d = cdist(x1, x2)
    
    K = varSigma*np.exp(-np.power(d, 2)/lengthscale)
    return K

def lin_kernel(x1, x2, varSigma):
    if x2 is None:
        return varSigma*x1.dot(x1.T)
    else:
        return varSigma*x1.dot(x2.T)
    
def white_kernel(x1, x2, varSigma):
    if x2 is None:
        return varSigma*np.eye(x1.shape[0])
    else:
        return np.zeros(x1.shape[0], x2.shape[0])
    
def periodic_kernel(x1, x2, varSigma, period, lengthScale):
    if x2 is None:
        d = cdist(x1, x1)
    else:
        d = cdist(x1, x2)
    return varSigma*np.exp(-(2*np.sin((np.pi/period)*d)**2)/lengthScale**2)
```

The kernels
The imshow plot shows the relative size of the values in the matrix. The diagonal terms from top left to bottom right show the size of the variance for a value with itself. The off-diagonal terms give the co-variance of each data point with every other data point. 

- **Radial Basis Function Kernel** The highest values are on the variance terms, given the distance function these are set to 1 * varSigma. The off diagnal covariance terms decrease the further two data points are from each other. This decrease is exponential, and the drop off can be slowed by increasing the length scale. Functions sampled from this will show correlated values at proximal points on the real number line with decreasing correlation for points further away. 


- **The Linear Kernel** The linear kernel shows the highest values of variance at the extremes of the distribution, where x multiplied by itself is the largest value. The center of the distribution shows the lowest variance where the variance will be 0. This indicates that all functions generated from this covariance matrix will pass through the mean value at that point. The opposite extremes of the distribution show negative covariance. Indicating that if one value is high the other will be low, this is a symmetric pattern and the reason that linear lines are sampled from the functions. 

- **The White Noise Kernel** Adds a specified amount of noise to the variance. Useful for approximating functions that don't pass exactly through the mean.

- **The Periodic Kernel** Covariance between set tuples of X points across the diagonals, indicating that patterns repeat across the x distribution, altering the period alters the frequency of covariance between tuples. This will restrict to functions which exhibit the same f(xi) at evenly spaced repetitions of xi.

**Question 4** Explain how we reach the updated belief about our function given the data? What is the specific process that we have to go through? Explain this both for any general model and what is unique (rather unique actually) about how we do this using a Gaussian process prior?

**Answer** In probability theory it is stated that the posterior probability is: *The probability of a specific distribution given the data*, is proportional to the multiplication of the *The probability of the data given a specific distribution* (likelihood function) multiplied by *The probability of the distribution* (prior), the prior and posterior must be conjugate distributions, such that when multiplied by the likelihood function give rise to each other, to allow for solving the posterior in closed form. In a general model we specify a prior distribution that we expect fits the distribution of the data, and compute the most likely parameters of the distribution having seen the data. In essence for each data point we correct the parameters of the distribution. 
A gaussian process is a parameterless prior, where we assume the function can be modelled by an infinite dimensional gaussian, and we restrict the prior to certain function spaces using the marginal properties of gaussians. That is the behaviour of any susbset of gaussians is equal to the subset of its components. Since the marginal property of gaussians states that any sample drawn from a gaussian is distributed according to a gaussian, we know the posterior must also be gaussian. The function space is restricted by modelling the covariance of f(xi) where f(xi) is drawn from the true distribution and the covariance between f(xi) is used to restrict the gaussian process to functions which pass through the true points.

**Answer** The predictive posterior mean is given by $K(X_{train}, X_{pred}) \cdot K(X_{train}, X_{train})^{-1} \cdot Y_{train}$ This can be thought of as generating a weights for the relationship between $X_{train}$ and $Y_{train}$, then using those weights to extrapolate at $X_{pred}$ based on the distance between $X_{train}$ and $X_{pred}$. 

The predictive posterior covariance is given by

$K(X_{pred}, X_{pred})$ - This can be thought of as the prior over the functions of the predicted values

$K(X_{pred}, X_{train})$ - This is the similarity of the two covariance matrices

$K(X_{train}, X_{train})^{-1}$ - This is the distance of the data with itself, in essence highly correlated points are inverted as they provide low information

In essence, the distance $X_{test}$ from $X_{train}$ determines the value of $K(X_{pred}, X_{train})$, if dist is high, value is small, correction to prior is small, i.e current data uninformative for predicted points, and inverse also true. 

The invers $X_{test}, X_{test}$ gives high values to points with low variance, which increases the correction to the prior for points well constrained by the training data, and accounts for redudency in information, i.e. two highly correlated points that are proximal in space say little about distant points, whereas points at a similar distance will have more impact on determining corrections to the posterior. 

Now let us implement the predictive posterior evaluated at data `x1, y1` and evaluated at `xstar`. You might want to be a bit more clever and implement a generic function that takes the matrices as input instead so that you can try different co-variance functions.



```python
def gp_posterior(x_train, y_train, x_test, kernel, lengthScale, varSigma, noise=1e-6): 
    
    K = kernel(x_train, x_train, varSigma, lengthScale) + white_kernel(x_train, None, noise)

    K_s = kernel(x_train, x_test, varSigma, lengthScale) 

    K_ss = kernel(x_test, x_test, varSigma, lengthScale)

    K_inv = np.linalg.inv(K)

    mu_s = K_s.T @ (K_inv @ y_train)

    cov_s = K_ss - K_s.T @ K_inv @ K_s

    return mu_s, cov_s, K, K_ss, K_s, K_inv  
    
```

```python
lengthScale = 20
varSigma = 1
N = 5
x = np.linspace(-3.1,4,N)
y = np.sin(2*np.pi/x) + x*0.1 + 0.3*np.random.randn(x.shape[0])
x1 = np.reshape(x,(-1,1))
y1 = np.reshape(y,(-1,1))
x_star = np.linspace(-6, 6, 500).reshape(-1, 1)
noise = 0.001

Nsamp = 100
mu_s, cov_s, K, K_ss, K_s, K_inv = gp_posterior(x1, y1, x_star, rbf_kernel, lengthScale, varSigma, noise=0.01)

fstar = np.random.multivariate_normal(mu_s.flatten(), cov_s, Nsamp)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x_star, fstar.T)
ax.scatter(x1, y1, 200, 'k', '*', zorder=2)
```

\begin{align}
p(\mathbf{f}_*\vert \mathbf{y}, \mathbf{x},\mathbf{x}_*,\theta) &= \mathcal{N}(\mu_{\mathbf{x}_*\vert \mathbf{x}}, K_{\mathbf{x}_*\vert \mathbf{x}})\\
\mu_{\mathbf{x}_*\vert \mathbf{x}} &= k(\mathbf{x}_*,\mathbf{x})\left(k(\mathbf{x},\mathbf{x})+\frac{1}{\beta}\mathbf{I}\right)^{-1}\mathbf{y}\\
K_{\mathbf{x}_*\vert \mathbf{x}} &= k(\mathbf{x}_*,\mathbf{x}_*) - k(\mathbf{x}_*,\mathbf{x})\left(k(\mathbf{x},\mathbf{x})+\frac{1}{\beta}\mathbf{I}\right)^{-1}k(\mathbf{x},\mathbf{x}_*).
\end{align}

In this final part we will go a bit beyond the material that we have covered in the lectures. So far we have specified the Gaussian process prior, performed the marginalisation and computed the predictive posterior. In this case the hyper-parameters, the parameters of the co-variance function have been fixed. We will now proceed to optimise the marginal likelihood as a way of "learning" the parameters. This means that our learning task is to,
$$
\hat{\boldsymbol{\theta}} = \operatorname{argmax}_{\boldsymbol{\theta}} p(\mathbf{y}\mid\mathbf{X},\boldsymbol{\theta}).
$$
Now in order to make things a little bit nicer for us to optimise we are instead of looking at the likelihood going to use the negative log-marginal likelihood instead and try to find the parameters that minimises this,
$$
\hat{\boldsymbol{\theta}} = \operatorname{argmin}_{\boldsymbol{\theta}} - \textrm{log}\left( p(\mathbf{y}\mid\mathbf{X},\boldsymbol{\theta})\right) = \mathcal{L}(\boldsymbol{\theta})
$$

**Question 8** Explain why these two objective functions lead to the same optimisation problem?

**Answer** The product of probabilities is equal to the sum of the log of probabilities. Maximising the probability, i.e. finding the maximum value for a probability is equivalent to finding the minimum value for the negative probability.

Now lets write up the log-marginal likelihood,
$$
\operatorname{log} p(\mathbf{y}\mid \mathbf{X}, \boldsymbol{\theta}) = \int p(\mathbf{y}\mid \mathbf{f})p(\mathbf{f}\mid \mathbf{X},\boldsymbol{\theta})\textrm{d}\boldsymbol{\theta}
$$
$$
-\frac{1}{2}\mathbf{y}^{\textrm{T}}\left(k(\mathbf{x},\mathbf{x})\right)^{-1}\mathbf{y} - \frac{1}{2}\operatorname{log}\operatorname{det}\left(k(\mathbf{x},\mathbf{x}-\beta^{-1}\mathbf{I}\right) - \frac{N}{2}\operatorname{log}2\pi = -\mathcal{L}(\boldsymbol{\theta})
$$

We will perform the optimisation using a gradient based optimiser. This means that we need to compute the gradients of the loss-function $\mathcal{L}$ with respect to $\boldsymbol{\theta}$. While these gradients are not very hard to compute by hand, and you are more than free to do so, we can also use packages that can handle `auto-differentiation`. There are lots of them around but my preference is to use `jax` which nicely follows `numpy`. Below is some scaffolding to a simple gradient based optimisation loop.



```python
from jax import grad
import jax.numpy as jnp

N = 5
x = np.linspace(-3.1,4,N)
y = np.sin(2*np.pi/x) + x*0.1 + 0.3*np.random.randn(x.shape[0])

x1 = np.reshape(x,(-1,1))
y1 = np.reshape(y,(-1,1))
y1 = y1 + np.random.normal(0, 2, size=y1.shape)
x_star = np.linspace(-6, 6, 500).reshape(-1, 1)

def squared_exponential(x1, x2, theta):
    """
    Squared Exponential Kernel (RBF) function.
    
    Parameters:
    - theta[0] : variance (signal variance)
    - theta[1] : lengthscale
    - theta[2] : noise variance (beta^{-1})
    """
    if x2 is None:
        # Compute pairwise squared Euclidean distance for the same inputs
        sqdist = jnp.sum(x1**2, axis=1).reshape(-1, 1) + jnp.sum(x1**2, axis=1) - 2 * jnp.dot(x1, x1.T)
        return theta[0] * jnp.exp(-0.5 / theta[1]**2 * sqdist) + 3 * jnp.eye(x1.shape[0])
    else:
        # Compute pairwise squared Euclidean distance between x1 and x2
        sqdist = jnp.sum(x1**2, axis=1).reshape(-1, 1) + jnp.sum(x2**2, axis=1) - 2 * jnp.dot(x1, x2.T)
        return theta[0] * jnp.exp(-0.5 / theta[1]**2 * sqdist)

def logmarginal_likelihood(theta, x, y):

    N = x.shape[0]  # Number of training points
    
    K = squared_exponential(x, None, theta)
    
    L = jnp.linalg.inv(K)

    part_one = -0.5 * jnp.dot(y.T, L) @ y.squeeze()
    
    part_two = -jnp.sum(jnp.log(jnp.diag(L)))

    part_three = -0.5 * N * jnp.log(2 * jnp.pi)
    
    return -(part_one + part_two + part_three)

dLdtheta = grad(logmarginal_likelihood)

theta = jnp.array([1.0, 1.0, 0.01])

for i in range(1000):
    grad_theta = dLdtheta(theta, x1, y)
    theta -= grad_theta * 0.01   
    
    if i % 100 == 0:
        print(theta)

print(theta)
varSigma, lengthScale, noise = theta
```
