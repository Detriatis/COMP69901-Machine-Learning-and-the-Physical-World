```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.spatial.distance import cdist
from sklearn.gaussian_process.kernels import RBF
from scipy.optimize import minimize
```

```python
def rbf_kernel(x1, x2, lengthScale, varSigma): 
    if x2 is None: 
        d = cdist(x1, x1) 
    else:
        d = cdist(x1, x2) 

    K = varSigma * np.exp(-np.power(d, 2) / lengthScale)
    
    return K 

def periodic_kernel(x1, x2, lengthScale, varSigma, period): 
    if x2 is None: 
        d = cdist(x1, x1) 
    else:
        d = cdist(x1, x2)

    K = varSigma * np.exp(-2 * (np.sin(np.pi * np.abs(d) / period) ** 2) / lengthScale**2)
    return K 

def lin_kernel(x1, x2, varSigma):
    if x2 is None:
        return varSigma*x1.dot(x1.T)
    else:
        return varSigma*x1.dot(x2.T)
```

```python
def gp_prior(x, theta): 
    N = x.shape[0] 
    
    mu = np.zeros(N)
    K = theta[0](x, None, theta[1], theta[2]) + theta[3] * np.eye(N)

    return mu, K 


def gp_posterior(x_train, y_train, x_test, kernel, lengthScale, varSigma, noise=1e-6): 
    
    K = kernel(x_train, x_train, lengthScale, varSigma) + noise * np.eye(len(x_train)) + lin_kernel(x_train, x_train, varSigma)

    K_s = kernel(x_train, x_test, lengthScale, varSigma) + lin_kernel(x_train, x_test, varSigma)

    K_ss = kernel(x_test, x_test, lengthScale, varSigma) + lin_kernel(x_test, x_test, varSigma)

    K_inv = np.linalg.inv(K)

    mu_s = K_s.T @ (K_inv @ y_train)

    cov_s = K_ss - K_s.T @ K_inv @ K_s

    return mu_s, cov_s 

def negative_log_likelihood(x_train, y_train, kernel, lengthScale, varSigma, noise=1e-6):
    K = kernel(x_train, x_train, lengthScale, varSigma)
    K = K + noise ** 2 * np.eye(len(x_train))
    D = K.shape[0]

    return  0.5 * D * np.log(2 * np.pi) + \
            0.5 * np.log(np.linalg.det(K)) + \
            0.5 * y_train.T @ np.linalg.inv(K) @ y_train 

from scipy.stats import norm 

def expected_improvement(f_max, mu, varSigma, x):

    improvement = f_max - mu 
    improvement_cdf = norm.cdf((improvement) / np.sqrt(varSigma.diagonal()))
    
    exploitation = improvement * improvement_cdf

    probability = norm.pdf(f_max, loc=mu, scale=varSigma.diagonal())
    exploration = varSigma.diagonal() * probability

    alpha = exploitation + exploration  

    return alpha, improvement_cdf, exploration 
```

```python
plt.close()
N = 100
N_train = 10
noise = 0.2


def sudo_function(x, beta=0.4, alpha1=1.0, alpha2=1.0):
    return np.sin(3.0*x) - alpha1*x + alpha2*x**2 + beta*np.random.randn(x.shape[0])

X = np.linspace(0, 10, N).reshape(N, 1) 
Y = sudo_function(X[:,0])

kernel = rbf_kernel
lengthScale = 5
varSigma = 1


#res = minimize(negative_log_likelihood, x0=theta, args=(X_train, Y_train, rbf_kernel, noise), method='L-BFGS-B')
#l_opt, sigma_f_opt = res.x
#print(f'lengthScale optimized {l_opt}\nvarSigma optimized: {sigma_f_opt}')

theta = [rbf_kernel, lengthScale, varSigma, noise]

mu, K = gp_prior(X, theta) 



for i in range(10): 
    if i == 0: 
        
        bit_mask = np.zeros(shape=X.shape[0], dtype=bool)
        max_indices = np.random.choice(X.shape[0], size=5, replace=False)
        

        Dataset = X.copy() 
        
        bit_mask[max_indices] = True 
        f_max = min(Y[bit_mask])  
    
    else: 
        
        mu_s, K_s = gp_posterior(X[bit_mask], Y[bit_mask], X, *theta)  
        
        try: 
            alphas, improvement_cdf, exploration  = expected_improvement(f_max, mu_s, K_s, X)
        except: 
            raise  

        
        max_indices = alphas == np.min(alphas[~bit_mask][np.nonzero(alphas[~bit_mask])])

       
        f_max = Y[max_indices]
       
        bit_mask[max_indices] = True 
       
        
        function_samples = np.random.multivariate_normal(mu_s, K_s, size=(50))
    
        fig, ax = plt.subplots(1, 1)
        
        ax.plot(Dataset, function_samples.T)
        
        ax.plot(Dataset, mu_s, color='black')
        
        ax.scatter(X[bit_mask], Y[bit_mask], zorder=2, color='black')

ax.scatter(X, Y) 



```

```python
def f(x, alpha=1.0, beta=0.5, gamma=0.2):
       return alpha*np.sin(2*np.pi*x/10) + ((beta*np.sin((2*np.pi*x/0.5))) +
                                            (gamma*x) +
                                            (0.1*np.random.randn(x.shape[0])))

kernel_cmpnd = ((GPy.kern.RBF(input_dim=1, variance=1.0, lengthscale=4.0)) +
                   (GPy.kern.StdPeriodic(input_dim=1, variance=1.0, period=10.0)) +
                   (GPy.kern.StdPeriodic(input_dim=1, variance=1.0, period=0.5)) +
                   (GPy.kern.Linear(input_dim = 1)) +
                   (GPy.kern.White(input_dim=1, variance=0.1)))

domain = [{'name': 'var_1', 'type': 'continuous', 'domain': (0,10)}]


opt = GPyOpt.methods.BayesianOptimization(f=f, domain=domain,model_type='GP',
                              kernel=kernel_cmpnd,
                              acquisition_type='EI',
                              initial_design_numdata=5)
opt.run_optimization(max_iter=10)
opt.plot_acquisition()
```
